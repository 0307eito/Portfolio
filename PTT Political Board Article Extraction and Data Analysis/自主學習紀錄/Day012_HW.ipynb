### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text,d.find(target="_blank").text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python

for e in soup.find(class_="part_list_2").find_all('h3'):
    print(e.find('em').text,' : ',e.find(target="_blank").text)
    
```

### ③ 哪一個類別的文章最多

```python
tag_all=[]

for f in soup.find(class_="part_list_2").find_all('em'):
    tag_all.append(f.text)
    
tag=(list(set(tag_all)))

for i in range(len(tag)):
    print(tag[i],':',tag_all.count(tag[i]))

print('類別為社會最多')
```
