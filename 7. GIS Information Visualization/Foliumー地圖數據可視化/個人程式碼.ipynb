```python
import openpyxl
wb = openpyxl.load_workbook("/Users/user/Desktop/汙染源.xlsx") #空氣污染、水污染、廢棄物、毒性化學物質等
ws = wb["工作表1"]
```

```python
print(ws["G2"].value is None)
```

```python
manufacturers_name=[]
manufacturers_address=[]
manufacturers_area_address=[]
pollution_type=[]
longitude=[]
latitude=[]
count=1000

for i in range(2,count+1):
    if ws['G'+str(i)].value and ws['D'+str(i)].value is not None:
        longitude.append(ws['G'+str(i)].value)
        latitude.append(ws['H'+str(i)].value)
        manufacturers_name.append(ws['A'+str(i)].value)
        manufacturers_address.append(ws['D'+str(i)].value)
        
        
        if ws['E'+str(i)].value is None:
            manufacturers_area_address.append("N/A")
        else:
            manufacturers_area_address.append(ws['E'+str(i)].value)
        
        if ws['F'+str(i)].value is None:
            pollution_type.append("N/A")
        else:
            pollution_type.append(ws['F'+str(i)].value)
            
        
        
        
```

```python
len(pollution_type)
```

```python
len(manufacturers_area_address)
```

```python
longitude_latitude=[]

for i in range(len(longitude)):
    longitude_latitude.append([latitude[i],longitude[i]])
```

```python
### 結合地圖做呈現
```

```python
import folium
from folium.plugins import MarkerCluster
from folium.plugins import HeatMap
```

```python


map = folium.Map(location=[25.043402, 121.615123], zoom_start=5)

for i in range(len(manufacturers_name)):
    
    FacilityName = "FacilityName: " + manufacturers_name[i]
    Address = "Address: " + manufacturers_address[i]
    IndustryArea = "IndustryArea: " + manufacturers_area_address[i]
    IndustryType = "IndustryType: " + pollution_type[i]
    
    folium.Marker(
        location=longitude_latitude[i],
        popup=folium.Popup(FacilityName+'<br>'+Address+'<br>'+IndustryArea+'<br>'+IndustryType,max_width=450),
        icon=folium.Icon(color='blue')
    ).add_to(map)

map
```

```python
map.save('Map.html')
```
