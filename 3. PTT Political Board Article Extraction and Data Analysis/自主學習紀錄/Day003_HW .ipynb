### 根據範例檔案的結果：

```python
# 1. 請問高雄市有多少地區有溫度資料？

with open('C:/Users/user/Downloads/zip/64_Weekday_CH.xml',encoding = 'utf8') as fd:
    d = dict(xmltodict.parse(fd.read()))
title = d['cwbopendata']['dataset']['datasetInfo']['datasetDescription']
locations=d['cwbopendata']['dataset']['locations']['locationsName']

data = d['cwbopendata']['dataset']['locations']['location']
for i in range(len(data)):
    print(data[i]['locationName']) 
print()
print('共'+str(len(data))+'個地區')
```

```python
# 2. 請取出每一個地區所記錄的第一個時間點跟溫度

#--------------------------------------------------------------------------------------------

locate = d['cwbopendata']['dataset']['locations']['location']
location=[]
for i in range(len(locate)):
    location.append(locate[i]['locationName'])
    

#--------------------------------------------------------------------------------------------

measures=[]

for i in range(0,38):
    time = d['cwbopendata']['dataset']['locations']['location'][i]['weatherElement']
    measures.append(time[0]['time'][0]['elementValue']['value'])
    

#--------------------------------------------------------------------------------------  

start_time=[]    

for i in range(0,38):
    time = d['cwbopendata']['dataset']['locations']['location'][i]['weatherElement']
    start_time.append(time[0]['time'][0]['startTime'])
    

#----------------------------------------------------------------------------------------

end_time=[]

for i in range(0,38):
    time = d['cwbopendata']['dataset']['locations']['location'][i]['weatherElement']
    end_time.append(time[0]['time'][0]['endTime'])


#-----------------------------------------------------------------------------------------

x=[]
for i in range(0,38):
    print(location[i])
    print(start_time[i])
    print(end_time[i])
    print('溫度'+measures[i]+'度')
    print()





```
