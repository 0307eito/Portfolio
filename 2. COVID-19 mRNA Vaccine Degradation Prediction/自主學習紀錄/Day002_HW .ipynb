```python
# 1. 取出班次一的每一個時間
x=[]
with open('C:/Users/user/Downloads/example.csv',encoding = 'utf8', newline='') as csvfile:
    rows = csv.reader(csvfile)
    for row in rows:
        x.append(row)
        
      
        
for i in range(len(x)):
    print(x[i][5],end='   ')
    
   
```

```python
# 2. 將班次一的每一個時間用一種資料型態保存
car=[]
dic={}
for i in range(len(x)):
    car.append(x[i][5])
    
car1=car[1:]


for i in range(len(x)-1):
    dic['1-'+str(i+1)]=car1[i]
    
print(dic)
    
```
